```python
%matplotlib inline

import numpy as np

from pantherapy.panthera import CrossSection

roughness = 0.035

z, y = np.loadtxt('data/stlms/xs.csv', delimiter=',', skiprows=1, unpack=True)

datum = 379.58
y = y - datum

sect = CrossSection(y, z, roughness)
_ = sect.plot(0)
```

```python
from dynrat.dynmod import Frict, QSolve


frict = Frict([-np.inf, np.inf], [0.035, 0.035])
bed_slope = 0.00011
slope_ratio = 40.40881261
time_step = 60*60

q_solve = QSolve(sect, frict, bed_slope, slope_ratio, time_step)
```

```python
h = 26.78
h_prime = 26.79
q_prime = 439000
q_solve.q_solve(h, h_prime, q_prime)
```
