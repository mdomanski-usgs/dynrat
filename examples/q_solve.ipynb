```python
%matplotlib inline

import numpy as np

from anchovy.crosssection import CrossSection

roughness = 0.035

station, elevation = np.loadtxt('data/stlms/xs.csv', delimiter=',', skiprows=1, unpack=True)

datum = 379.58
elevation = elevation - datum

sect = CrossSection(station, elevation, roughness)
_ = sect.plot(0)
```

```python
from dynrat.dynmod import QSolver
from dynrat.frict import TableFrict


frict = TableFrict([-np.inf, np.inf], [0.035, 0.035])
bed_slope = 0.00011
slope_ratio = 40.40881261
time_step = 60*60

q_solver = QSolver(sect, frict, bed_slope, slope_ratio, time_step)
```

```python
h = 26.78
h_prime = 26.79
q_prime = 439000
q_solver.q_solve(h, h_prime, q_prime)
```
